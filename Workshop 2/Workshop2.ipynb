```python
mean_survival = df2.groupby(['Pclass', 'Age'])['Survived'].mean().reset_index()

# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(x='Pclass', y='Survived', hue='Age', data=mean_survival)

# Customizing the plot
plt.title('Survival Rate by Class and Age Group', fontsize=16)
plt.xlabel('Class (Pclass)', fontsize=12)
plt.ylabel('Mean Survival Rate', fontsize=12)
plt.legend(title='Age Group', loc='upper left')

plt.show()
```
